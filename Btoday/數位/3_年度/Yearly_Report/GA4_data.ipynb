## 將各月份GA4報表資料合併
每月份的資料中，可能會有單篇文章出現在各個不同月份的情況，因此需要將相同ID的文章瀏覽量做加總。
- 原始csv檔案格式有做調整，需要複製前一月的csv，再從此月份的xlsx檔貼上資料。否則會出現檔案讀取問題

```python
import pandas as pd
import glob
```

### 讀取資料

```python
read_files = glob.glob(r'//192.168.0.26/數位內容區/行銷組/數據分析資料/每月提供/每月資料/今周刊/GA4/csv/*.csv')

print(read_files)
```

### 合併dataframe

```python
df_list = []

for file in read_files:
    df_csv = pd.read_csv(f"{file}",  encoding = "UTF-8-SIG", skipinitialspace = True)#, na_values = ['null'])
    
    df_list.append(df_csv)
#     df_3 = [df_1, df_2]

df_all = pd.concat(df_list, ignore_index = True) 
    
print(df_all)
```

```python
df_all.columns.values
```

```python
df_unique = df_all.drop("瀏覽量", axis = 1)

df_unique = df_unique.drop_duplicates(ignore_index = True)

print(df_unique)
```

```python
column_names = ['文章ID']
```

```python
summeries = {'瀏覽量': 'sum'}

df_all_sum = df_all.groupby(by = column_names).agg(summeries)

df_all_sum = df_all_sum.sort_values("瀏覽量", ascending = False).reset_index()

df_all_sum = df_all_sum.rename(columns={"瀏覽量": "瀏覽量_Sum"})

df_all_sum_all = pd.merge(df_all_sum, df_unique, how = 'inner')
# print(df_all_sum)

df_all_sum_all.head()
```

```python
len(df_all_sum_all)
```

```python
df_all_sum_all = df_all_sum_all.drop_duplicates(subset = ["文章ID"], keep = "last").reset_index(drop = True)

df_all_sum_all.head(10)
```

```python
len(df_all_sum_all)
```

```python
df_all_sum_all.to_csv("./2023/ALL_Data/GA4_ALL_DATA.csv", index = False, encoding = 'utf-8-sig')
```
