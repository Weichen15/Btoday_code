```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

## Read data

```python
df_user = pd.read_csv("GA4_DAY_AGE_GENDER_2023.csv")

df_user_bk = df_user.copy()
```

```python
df_user.head()
```

### Date(Day of Week) 改成時間資料

```python
# 將星期數字轉換為時間型態
start_date = datetime(2023, 1, 1)  # 使用一個基準日期，這裡假設是2023年1月1日
```

```python
df_user["Day"] = df_user["Date (Day of Week)"].apply(lambda x: (start_date + timedelta(days = x)).strftime("%A"))
```

```python
df_user.head()
```

```python
df_user_bk_2 = df_user.copy()
```

```python
df_user.drop(["Date (Day of Week)"], axis = 1, inplace = True)
```

```python
df_user.head()
```

```python
data_age = df_user.groupby(["Day", "Age"]).sum()
```

```python
data_gender = df_user.groupby(["Day", "Gender"]).sum()
```

```python
data_age_gender = df_user.groupby(["Day", "Age", "Gender"]).sum()
```
