```python
import pandas as pd
import numpy as np
from datetime import datetime
```

## Read data

```python
df_user = pd.read_csv("GA4_AGE_CHANNEL_GROUP_2023.csv")

df_user_bk = df_user.copy()
```

```python
df_user.head()
```

```python
df_user["year_month"] = df_user["Date (Year Month)"].apply(lambda x: datetime.strptime(x, "%b %Y").strftime("%Y/%m"))
# df_user["year_month"] = df_user["Date (Year Month)"].apply(lambda x: datetime.strptime(x, "%b %Y"))
```

```python
df_user.head()
```

```python
df_organic_serach = df_user[df_user["Default channel group"] == "Organic Search"]
df_organic_serach.head()
```

```python
data_organic_serach = df_organic_serach.groupby(["year_month", "Age"]).sum()
```

```python
data_organic_serach
```

```python
data_organic_serach.to_csv("./data/2023_ga4_user_age_organic_search.csv")
```

```python
df_organic_social = df_user[df_user["Default channel group"] == "Organic Social"]
df_organic_social.head()
```

```python
data_organic_social = df_organic_social.groupby(["year_month", "Age"]).sum()
data_organic_social.head()
```

```python
data_organic_social.to_csv("./data/2023_ga4_user_age_organic_social.csv")
```
