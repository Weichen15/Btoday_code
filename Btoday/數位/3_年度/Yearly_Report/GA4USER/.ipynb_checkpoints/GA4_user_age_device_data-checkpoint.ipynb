```python
import pandas as pd
import numpy as np
from datetime import datetime
```

## Read data

```python
df_user = pd.read_csv("GA4_AGE_DEVICE_2023.csv")

df_user_bk = df_user.copy()
```

```python
df_user.head()
```

```python
df_user["year_month"] = df_user["Date (Year Month)"].apply(lambda x: datetime.strptime(x, "%b %Y").strftime("%Y/%m"))
# df_user["year_month"] = df_user["Date (Year Month)"].apply(lambda x: datetime.strptime(x, "%b %Y"))
```

```python
df_user.head()
```

```python
data_device = df_user.groupby(["year_month", "Device category"]).sum()
```

```python
data_device
```

```python
data_age_device_1 = df_user.groupby(["Age", "Device category"]).sum()
```

```python
data_age_device_1
```

```python
data_age_device_2 = df_user.groupby(["year_month", "Age", "Device category"]).sum()
```

```python
data_age_device_2
```

```python
data_device.to_csv("./data/2023_ga4_user_device.csv")
```

```python
data_age_device_1.to_csv("./data/2023_ga4_user_age_device.csv")
```
