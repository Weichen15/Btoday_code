```python
import pandas as pd
import numpy as np
from datetime import datetime
```

## Read data

```python
df_user = pd.read_csv("GA4_AGE_GENDER_2023.csv")

df_user_bk = df_user.copy()
```

```python
df_user.head()
```

```python
df_user["year_month"] = df_user["Date (Year Month)"].apply(lambda x: datetime.strptime(x, "%b %Y").strftime("%Y/%m"))
# df_user["year_month"] = df_user["Date (Year Month)"].apply(lambda x: datetime.strptime(x, "%b %Y"))
```

```python
df_user.head()
```

```python
data_age = df_user.groupby(["year_month", "Age"]).sum()
```

```python
data_age
```

```python
data_gender = df_user.groupby(["year_month", "Gender"]).sum()
```

```python
data_gender
```

```python
data_age_gender = df_user.groupby(["year_month", "Age", "Gender"]).sum()
```

```python
data_age_gender
```

```python
data_age.to_csv("./data/2023_ga4_user_age.csv")
```

```python
data_gender.to_csv("./data/2023_ga4_user_gender.csv")
```

```python
data_age_gender.to_csv("./data/2023_ga4_user_age_gender.csv")
```
