```python
import numpy as np
import pandas as pd
import re

from datetime import datetime, timedelta
```

```python
file_name = input("Please enter date range: ")
```

```python
df_data = pd.read_excel(f"../data/Ga4報表_今周刊_25000_{file_name}.xlsx")

df_data.head()
```

## 1. 抓取當周的文章

```python
df_data["文章ID"] = df_data["文章ID"].astype(np.str)

df_data["article_date"] = df_data["文章ID"].apply(lambda X: X[:8])

df_data["article_date"] = pd.to_datetime(df_data["article_date"])

df_data["article_date"].head()
```

```python
specified_date = datetime(2024, 2, 12)

last_monday = specified_date - timedelta(days = 7)

last_sunday = specified_date - timedelta(days = 1)

print("上周一的日期：", last_monday.strftime("%Y-%m-%d"))
print("上周日的日期：", last_sunday.strftime("%Y-%m-%d"))
```

```python
df_weekly_data = df_data[last_monday <= df_data["article_date"]]
```

```python
df_weekly_data = df_weekly_data[df_data["article_date"] <= last_sunday]
```

```python
df_weekly_data.head()
```

```python
df_weekly_data["article_date"].value_counts().sort_index()
```

## 2. 篩選文章來源

```python
article_source = ["今周/特刊/叢書"]

df_weekly_data = df_weekly_data[~df_weekly_data["文章來源"].isin(article_source)].reset_index(drop = True)
```

```python
df_weekly_data.head()
```

```python
df_weekly_data["文章來源"].value_counts()
```

## 3. 計算各屬性文章篇數

```python
df_weekly_data["文章屬性"].value_counts()
```

```python
df_weekly_data[df_weekly_data["文章屬性"] == "投資理財"]
```

## 4. 當周SEO分類

```python
df_SEO = df_weekly_data[~df_weekly_data["SEO分類"].isnull()]
df_SEO.reset_index(inplace = True, drop = True)
```

```python
from collections import Counter
```

```python
# 將每個字詞區隔開來，並計算頻率
word_freq = Counter([word for row in df_SEO["SEO分類"] for word in row.split("/")])

# 將結果轉換為 DataFrame
df_word_freq = pd.DataFrame(list(word_freq.items()), columns = [ "SEOtag" , "次數"])

# 顯示 DataFrame
df_word_freq.sort_values(by = "次數", ascending = False, inplace = True)
df_word_freq.head(20)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
## 定義中文字體
my_font_path = "C:/Users/JoeJian/OneDrive - Business Today Magazine/桌面/Python Codes/Font/jf-openhuninn-2.0.ttf"

# 將標籤和次數整理成字典
tag_dict = dict(zip(df_word_freq["SEOtag"], df_word_freq["次數"]))

# 使用WordCloud生成文字雲
wordcloud = WordCloud(width = 800, height = 400, font_path = my_font_path, background_color = "white").generate_from_frequencies(tag_dict)

# 顯示生成的文字雲
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.show()
```
