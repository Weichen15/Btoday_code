# 分類每月編輯部數位稿

### 編輯部要統計每月數位稿的字數

## 1. import modules

```python
import pandas as pd
import numpy as np

import re
```

## 2. import data

### 下載下來的資料都有相同的命名格式，先輸入日期，選取欲整理的資料

```python
df_all = pd.read_csv(".\重複資料\GA報表_PV_sum_all_202301_202311.csv", encoding = "utf-8")

df_all.head()
```

## 3. 設定文章ID

```python
df_all["文章ID"] = df_all["文章ID"].astype("str")

df_all["文章ID"].head(10)
```

## 4. 選取編輯部的資料
### 文章來源：編輯部記者

```python
df_paper_editior = pd.DataFrame(columns = df_all.columns)

for num in range(len(df_all)):
    if df_all['文章來源'][num] == "編輯部記者":
        # 使用 loc 函數抓出符合條件的行並添加到新的 DataFrame 中
        selected_row = df_all.loc[num]
        df_paper_editior = df_paper_editior.append(selected_row, ignore_index = True)

# 顯示新的 DataFrame
df_paper_editior.head(10)
```

## 5. 從文章ID抓取各月份資料

```python
year_list = ["2023"]
month_list = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11"]#, "12"]
```

```python
for year_num in range(len(year_list)):
    for month_num in range(len(month_list)):
        df_paper_editor_monthly = pd.DataFrame(columns = df_paper_editior.columns)
        
        for num in range(len(df_paper_editior)):
            
            if re.match(f"{year_list[year_num]}{month_list[month_num]}+", df_paper_editior["文章ID"][num]):
                selected_row = df_paper_editior.loc[num]
                df_paper_editor_monthly = df_paper_editor_monthly.append(selected_row, ignore_index = True)
#             else:
#                 print('False')
        df_paper_editor_monthly.to_csv(f'編輯部記者資料_{year_list[year_num]}{month_list[month_num]}.csv', encoding = 'utf-8-sig', index = False)
        
        # Assuming you want to keep a list of article titles for each editor
        editor_stats = df_paper_editor_monthly.groupby('撰文').agg({'文章字數': 'sum', '文章標題': lambda x: list(x)}).reset_index()

        # The resulting DataFrame will have three columns: "editor", "page views", and "article titles"
        editor_stats.to_csv(f'編輯部記者資料_字數統計_{year_list[year_num]}{month_list[month_num]}.csv', encoding = 'utf-8-sig', index = False)
        print("year, month: ", year_list[year_num], month_list[month_num])

```
