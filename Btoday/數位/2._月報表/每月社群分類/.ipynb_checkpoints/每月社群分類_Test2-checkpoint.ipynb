# 分類每月社群報告文章組別

### 每月社群報告執行前，要先將下載下來的文章分類到對應的群組，由以下程式幫忙做大致分類，細項仍需要後續手動調整

## 1. import modules

```python
import pandas as pd
import numpy as np
```

## 2. import data

### 下載下來的資料都有相同的命名格式，先輸入日期，選取欲整理的資料

```python
date_range = input("請輸入月份區間(YYYYMMDD_YYYYMMDD): ")
```

```python
df_all = pd.read_csv(f"今周刊_{date_range}_整理.csv", encoding = 'utf-8')

df_all.head()
```

## 3. 設定各組別名單

```python
# 人員會有異動、需要定期更新名單

lst_class_other = ['今周/特刊/叢書', '編輯部記者', '本刊文章']
lst_reporters = ['林信男', '劉煥彥', '林韋伶', '廖元鈴', '黃健誠', '蕭婷方', '許雅綿', '蘇君薇', '彭蕙珍', '鄭鴻達', '鴻達']
lst_social_editors = ['王郁惠', '林可薇', '李伊晴', '林依榕', '翁申霖', '程子威', '張家麒', '李易紓', '胡肇芳', '肇芳', '黃明惠', '張若儀']
lst_blog_editors = ['洪慧如', '歐陽蓉', '胡詠晴']
lst_podcast_editors = ['潘怡靜', '張孝瑜', '張燕如', '盧沛甫', '諶儀璇']
lst_other_editors = ['林芷揚', '趙閎毅', '林暐鈞', '郭美懿']
lst_video_editors = ['莊翊晨', '王君瑭', '鍾子葳', '呂佩芬', '蔡銘洋', '蔡侑儒', '朱嘉淳', '林慧詩']
```

```python
# 用來存方文章id與文章分類

lst_article_id = []
lst_article_class = []
```

## 4. 定義分類Function

```python
def func_article_classification(df_name, iloc_num, editor_identification):
    """
    參數1帶入：DataFrame名稱
    參數2帶入：i
    參數3填入：「撰文」、「上稿者」或「最後修改者」
    """
    try:
        if df_name.iloc[iloc_num][f'{editor_identification}'][:3] in lst_reporters:

            if df_name.iloc[iloc_num][f'{editor_identification}'][:3] == '黃健誠':

                if df_name.iloc[iloc_num]['文章發布'] >= '2019-05-01':

                    lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
                    lst_article_class.append('記者')

                else:
                    lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
                    lst_article_class.append('社群')

            elif df_name.iloc[iloc_num][f'{editor_identification}'][:3] == '蘇君薇':

                if df_name.iloc[iloc_num]['文章發布'] >= '2020-02-01':
                    lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
                    lst_article_class.append('記者')

                else:
                    lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
                    lst_article_class.append('社群')

            else:

                lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
                lst_article_class.append('記者')

        elif df_name.iloc[iloc_num][f'{editor_identification}'][:3] in lst_social_editors:

            if df_name.iloc[iloc_num][f'{editor_identification}'][:3] == '李伊晴':

                if df_name.iloc[iloc_num]['文章發布'] >= '2020-06-01':

                    lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
                    lst_article_class.append('其他')

                else:
                    lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
                    lst_article_class.append('社群')

            elif df_name.iloc[iloc_num][f'{editor_identification}'][:3] == '張家麒':

                if df_name.iloc[iloc_num]['文章發布'] >= '2020-03-01':

                    lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
                    lst_article_class.append('社群')

                else:
                    lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
                    lst_article_class.append('部落客')

            elif df_name.iloc[iloc_num][f'{editor_identification}'][:3] == '李易紓':

                if df_name.iloc[iloc_num]['文章發布'] >= '2022-02-01':

                    lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
                    lst_article_class.append('社群')

                else:
                    lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
                    lst_article_class.append('其他')

            else:

                lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
                lst_article_class.append('社群')

        elif df_name.iloc[iloc_num][f'{editor_identification}'][:3] in lst_blog_editors:

            lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
            lst_article_class.append('部落客')

        elif df_name.iloc[iloc_num][f'{editor_identification}'][:3] in lst_podcast_editors:

            lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
            lst_article_class.append('podcast')
            
        elif df_name.iloc[iloc_num][f'{editor_identification}'][:3] in lst_other_editors:

            lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
            lst_article_class.append('其他')

        elif df_name.iloc[iloc_num][f'{editor_identification}'][:3] in lst_video_editors:

            lst_article_id.append(df_name.iloc[iloc_num]['文章ID'])
            lst_article_class.append('影音')
    except:
        pass
```

```python
### Way2

def func_editor_classification(df_name, iloc_num):
    """
    參數1帶入：DataFrame名稱
    參數2帶入：i
    """
    
    lst_editor_identification = ['撰文', '上稿者', '最後修改者']
    
    for editor_identification in lst_editor_identification:
        
        if editor_identification == '最後修改者':
            
            if str(df_all.iloc[iloc_num]['上稿者']) == str(np.nan):
                
                func_article_classification(df_name, iloc_num, lst_editor_identification)
        
        else:
            
            func_article_classification(df_name, iloc_num, lst_editor_identification)
        
        
```

## 5. 開始分類文章

```python
# Step 1. 先判斷其他

for i in range(len(df_all)):
#     print(df_all_top50.iloc[i]['文章來源'])

    if df_all.iloc[i]['文章來源'] in lst_class_other:
#         print(df_all_top50.iloc[i]['文章來源'])
        lst_article_id.append(df_all.iloc[i]['文章ID'])
        lst_article_class.append('其他')

    
# Step 2. 從撰文判斷分類

for i in range(len(df_all)):
    
    if df_all.iloc[i]['文章ID'] not in lst_article_id:
        
        func_article_classification(df_all, i, '撰文')

# Step 3. 從上稿者判斷分類

for i in range(len(df_all)):
    
    if df_all.iloc[i]['文章ID'] not in lst_article_id:
        
        func_article_classification(df_all, i, '上稿者')

# Step 4. 從上稿者為空白，從最後修改者去分類

for i in range(len(df_all)):
    
    if (df_all.iloc[i]['文章ID'] not in lst_article_id) and (str(df_all.iloc[i]['上稿者']) == str(np.nan)):
        
        func_article_classification(df_all, i, '最後修改者')

for i in range(len(df_all)):
    
    if df_all.iloc[i]['文章ID'] not in lst_article_id:
        
        lst_article_id.append(df_all.iloc[i]['文章ID'])
        lst_article_class.append('其他')
        

print("finish")
```

```python
### Way2

# Step 1. 先判斷其他

for i in range(len(df_all)):

    if df_all.iloc[i]['文章來源'] in lst_class_other:

        lst_article_id.append(df_all.iloc[i]['文章ID'])
        lst_article_class.append('其他')

    else:
        
        if df_all.iloc[i]['文章ID'] not in lst_article_id:

            func_editor_classification(df_all, i)


#             if df_all.iloc[i]['文章ID'] not in lst_article_id:
        
#                 lst_article_id.append(df_all.iloc[i]['文章ID'])
#                 lst_article_class.append('其他')
        

print("finish")
```

```python
df_class = pd.DataFrame({'文章ID': lst_article_id,
                         '文章分類': lst_article_class})
                         
df_class
```

```python
df_class_final = pd.merge(df_all, df_class, how = 'left')

df_class_final
```
