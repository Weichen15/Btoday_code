```python
# 需要定期更新

lst_class_other = ['今周/特刊/叢書', '編輯部記者', '本刊文章']
lst_reporters = ['林信男', '劉煥彥', '林韋伶', '廖元鈴', '黃健誠', '蕭婷方', '許雅綿', '蘇君薇', '彭蕙珍', '鄭鴻達', '鴻達']
lst_social_editors = ['王郁惠', '林可薇', '李伊晴', '林依榕', '翁申霖', '程子威', '張家麒', '李易紓', '胡肇芳', '肇芳', '黃明惠', '張若儀']
lst_blog_editors = ['洪慧如', '歐陽蓉', '胡詠晴']
lst_other_editors = ['林芷揚', '趙閎毅', '林暐鈞', '郭美懿', '潘怡靜', '張孝瑜', '張燕如', '盧沛甫', '諶儀璇']
lst_video_editors = ['莊翊晨', '王君瑭', '鍾子葳', '呂佩芬', '蔡銘洋', '蔡侑儒', '朱嘉淳', '林慧詩']
```

```python
lst_article_id = []
lst_article_class = []
```

```python
lst_id_count_temp = [] # 統計用
```

```python
df_all_top50 = df_all[:500]
```

```python
# Step 1. 先判斷其他

for i in range(len(df_all_top50)):
#     print(df_all_top50.iloc[i]['文章來源'])

    if df_all_top50.iloc[i]['文章來源'] in lst_class_other:
#         print(df_all_top50.iloc[i]['文章來源'])
        lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
        lst_article_class.append('其他')

    
# Step 2. 從撰文判斷分類

for i in range(len(df_all_top50)):
    
    if df_all_top50.iloc[i]['文章ID'] not in lst_article_id:
        
        if df_all_top50.iloc[i]['撰文'][:3] in lst_reporters:
            
            if df_all_top50.iloc[i]['撰文'][:3] == '黃健誠':
                
                if df_all_top50.iloc[i]['文章發布'] >= '2019/05/01':
                    
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('記者')
                    
                else:
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('社群')
                    
            elif df_all_top50.iloc[i]['撰文'][:3] == '蘇君薇':
                
                if df_all_top50.iloc[i]['文章發布'] >= '2020/02/01':
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('記者')
                    
                else:
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('社群')
                    
            else:
                
                lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                lst_article_class.append('記者')
        
        elif df_all_top50.iloc[i]['撰文'][:3] in lst_social_editors:
                        
            if df_all_top50.iloc[i]['撰文'][:3] == '李伊晴':
                
                if df_all_top50.iloc[i]['文章發布'] >= '2020/06/01':
                    
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('其他')
                    
                else:
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('社群')
                    
            elif df_all_top50.iloc[i]['撰文'][:3] == '張家麒':
                
                if df_all_top50.iloc[i]['文章發布'] >= '2020/03/01':
                    
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('社群')
                    
                else:
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('部落客')
                    
            elif df_all_top50.iloc[i]['撰文'][:3] == '李易紓':
                
                if df_all_top50.iloc[i]['文章發布'] >= '2022/02/01':
                    
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('社群')
                    
                else:
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('其他')
                    
            else:
                
                lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                lst_article_class.append('社群')
            
        elif df_all_top50.iloc[i]['撰文'][:3] in lst_blog_editors:

            lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
            lst_article_class.append('部落客')

        elif df_all_top50.iloc[i]['撰文'][:3] in lst_other_editors:

            lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
            lst_article_class.append('其他')

        elif df_all_top50.iloc[i]['撰文'][:3] in lst_video_editors:

            lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
            lst_article_class.append('影音')
            

# Step 3. 從上稿者判斷分類

for i in range(len(df_all_top50)):
    
    if df_all_top50.iloc[i]['文章ID'] not in lst_article_id:
        try:
            if df_all_top50.iloc[i]['上稿者'][:3] in lst_reporters:

                if df_all_top50.iloc[i]['上稿者'][:3] == '黃健誠':

                    if df_all_top50.iloc[i]['文章發布'] >= '2019/05/01':

                        lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                        lst_article_class.append('記者')

                    else:
                        lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                        lst_article_class.append('社群')

                elif df_all_top50.iloc[i]['上稿者'][:3] == '蘇君薇':

                    if df_all_top50.iloc[i]['文章發布'] >= '2020/02/01':
                        lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                        lst_article_class.append('記者')

                    else:
                        lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                        lst_article_class.append('社群')

                else:

                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('記者')

            elif df_all_top50.iloc[i]['上稿者'][:3] in lst_social_editors:

                if df_all_top50.iloc[i]['上稿者'][:3] == '李伊晴':

                    if df_all_top50.iloc[i]['文章發布'] >= '2020/06/01':

                        lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                        lst_article_class.append('其他')

                    else:
                        lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                        lst_article_class.append('社群')

                elif df_all_top50.iloc[i]['上稿者'][:3] == '張家麒':

                    if df_all_top50.iloc[i]['文章發布'] >= '2020/03/01':

                        lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                        lst_article_class.append('社群')

                    else:
                        lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                        lst_article_class.append('部落客')

                elif df_all_top50.iloc[i]['上稿者'][:3] == '李易紓':

                    if df_all_top50.iloc[i]['文章發布'] >= '2022/02/01':

                        lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                        lst_article_class.append('社群')

                    else:
                        lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                        lst_article_class.append('其他')
                else:
                
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('社群')

            elif df_all_top50.iloc[i]['上稿者'][:3] in lst_blog_editors:

                lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                lst_article_class.append('部落客')

            elif df_all_top50.iloc[i]['上稿者'][:3] in lst_other_editors:

                lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                lst_article_class.append('其他')

            elif df_all_top50.iloc[i]['上稿者'][:3] in lst_video_editors:

                lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                lst_article_class.append('影音')
        except:
            
            pass

# Step 4. 從上稿者為空白，從最後修改者去分類

for i in range(len(df_all_top50)):
    
    if (df_all_top50.iloc[i]['文章ID'] not in lst_article_id) and (str(df_all_top50.iloc[i]['上稿者']) == str(np.nan)):
        
        if df_all_top50.iloc[i]['最後修改者'][:3] in lst_reporters:
            
            if df_all_top50.iloc[i]['最後修改者'][:3] == '黃健誠':
                
                if df_all_top50.iloc[i]['文章發布'] >= '2019/05/01':
                    
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('記者')
                    
                else:
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('社群')
                    
            elif df_all_top50.iloc[i]['最後修改者'][:3] == '蘇君薇':
                
                if df_all_top50.iloc[i]['文章發布'] >= '2020/02/01':
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('記者')
                    
                else:
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('社群')
                    
            else:
                
                lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                lst_article_class.append('記者')
        
        elif df_all_top50.iloc[i]['最後修改者'][:3] in lst_social_editors:
            
            if df_all_top50.iloc[i]['最後修改者'][:3] == '李伊晴':
                
                if df_all_top50.iloc[i]['文章發布'] >= '2020/06/01':
                    
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('其他')
                    
                else:
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('社群')
                    
            elif df_all_top50.iloc[i]['最後修改者'][:3] == '張家麒':
                
                if df_all_top50.iloc[i]['文章發布'] >= '2020/03/01':
                    
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('社群')
                    
                else:
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('部落客')
                    
            elif df_all_top50.iloc[i]['最後修改者'][:3] == '李易紓':
                
                if df_all_top50.iloc[i]['文章發布'] >= '2022/02/01':
                    
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('社群')
                    
                else:
                    lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                    lst_article_class.append('其他')
                    
            else:
                
                lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
                lst_article_class.append('社群')
            
        elif df_all_top50.iloc[i]['最後修改者'][:3] in lst_blog_editors:

            lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
            lst_article_class.append('部落客')

        elif df_all_top50.iloc[i]['最後修改者'][:3] in lst_other_editors:

            lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
            lst_article_class.append('其他')

        elif df_all_top50.iloc[i]['最後修改者'][:3] in lst_video_editors:

            lst_article_id.append(df_all_top50.iloc[i]['文章ID'])
            lst_article_class.append('影音')
        

# print(len(lst_article_id), len(lst_id_count_temp))
```

```python
df_class = pd.DataFrame({'文章ID': lst_article_id,
                         '文章分類': lst_article_class})
                         
df_class
```

```python
df_class_final = pd.merge(df_all_top50, df_class, how = 'left')

df_class_final
```
