### Traffic欄位型態調整成int，才能做之後的計算

```python
## 先把欄位資料都改成str，才可以做replace
df_kw["Traffic "] = df_kw["Traffic "].astype(np.str)

## 用replace把千位數的","去除掉
df_kw["Traffic "] = df_kw["Traffic "].apply(lambda X: X.replace(",", ""))

## 用replace把小數點的"."去除掉
df_kw["Traffic "] = df_kw["Traffic "].apply(lambda X: X.replace(".", ""))

## 用replace把nan改成0
df_kw["Traffic "] = df_kw["Traffic "].apply(lambda X: X.replace("nan", "0"))
```

```python
## 成功後再將資料改為np.int64
df_kw["Traffic "] = df_kw["Traffic "].astype(np.int64)
```

### 輸入有幾組網站

```python
num_webs = input("Please enter how many websites of this data: ")
```

### 轉換百分比資料

```python
# 定義函示作資料轉換
def format_percentage(percentage):
    if pd.isna(percentage):  # 如果是 NaN，直接返回
        return "NaN"
    else:

        # 将所有非字符串的数据转换为字符串
        if not isinstance(percentage, str):
            percentage = str(percentage)
        
        # 如果是百分比字符串，去除百分号并转换为浮点数
        if '%' in percentage:
            percentage = float(percentage.strip('%'))
        # 如果是浮点数，乘以100
        elif '.' in percentage:
            percentage = float(percentage) * 100
        # 格式化为百分比形式字符串
        return '{:.{}f}%'.format(float(percentage), 2)  # 只保留一位小数
```

```python
percent_list = column_name[7: 7 + int(num_webs)]
print(percent_list)
```

```python
for i in range(len(percent_list)):
    df_kw[f"{percent_list[i]}"] = df_kw[f"{percent_list[i]}"].apply(format_percentage)
```

```python
df_share = pd.DataFrame()

for num_web in range(int(num_webs)):
    traffic_share = []
    company_name = column_name[num_web + 7].split(".")[0]
    
    print(company_name)
    
    df_kw[f"{column_name[num_web + 7]}"] = df_kw[f"{column_name[num_web + 7]}"].astype("str")
    
    df_share[f"{company_name}_share"] = df_kw[f"{column_name[num_web + 7]}"].str.replace('%', '').astype(float) / 100.0

    for num in range(len(df_kw)):
        
        traffic_share.append(df_kw["Traffic "][num] * df_share[f"{company_name}_share"][num])
    
    df_kw[f"{company_name}_traffic_share"] = traffic_share
print("--- finish ---")
```
