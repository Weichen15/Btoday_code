```python
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
# import date

url = "https://www.businesstoday.com.tw/article/category/183013/post/202304260033/"
headers={"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36"}
response = requests.get(url, headers = headers)
bt_html = BeautifulSoup(response.text, "lxml")
print(bt_html)
```

```python
# 大標題：00878正式配息公告！三個理由存股必買：就算19元高點入場也是賺！下週除息這天最後買進日
list_title_h1 = []
title1 = bt_html.find_all("h1")  # 搜尋所有h1
for title in title1:
    print(title.getText())  # 只提取文字部分
    list_title_h1.append(title.getText())
```

```python
list_title_h1
```

```python
# 次標題：
list_title_h2 = []
title2 = bt_html.find_all("h2")  # 搜尋所有h2
for title in title2:
    print(title.getText())  # 只提取文字部分
    list_title_h2.append(title.getText())
```

```python
list_title_h2
```

```python
content = bt_html.find_all("div", {"class":"article__wrap"})  # 文章內容
print(content)
```

```python
# 文章分類及文章發佈日
list_content_class = []
list_content_date = []

for ct in content:  # r: 一個li區塊
    ct_class = ct.find("p", {"class":"context__info-item context__info-item--type"})  # 文章分類
    ct_date  = ct.find("p", {"class":"context__info-item context__info-item--date"})  # 文章發佈日
    list_content_class.append(ct_class.getText())
    list_content_date.append(ct_date.getText())
    print(ct_class.getText())
    print(ct_date.getText())
```

```python
# 文章屬性

print(bt_html.find("meta", {"property":"article:section"}).get("content"))
```

```python
# instruction

list_content_instruction = []


for ct in content:  # r: 一個li區塊
    ct_introduction = ct.find("div", {"class":"article__introduction"})  # 文章簡介
    list_content_instruction.append(ct_introduction.getText())
    print(ct_introduction.getText())
    
```

```python
list_content_instruction
```

```python
# all content

list_content_all = []


for ct in content:  # r: 一個li區塊
    ct_all = ct.find("div", {"id":"article__page__202304260033"})  # 文章內文 (article+page+文章ID)
#     print(ct_all)
    list_content_all.append(ct_all.getText())
#     print(ct_all.getText())
    
```

```python
list_content_all
```

```python
# 標籤
list_seo_tag = []
SEOTAG = bt_html.find_all("a", {"class":"context__tag"})  # 標籤
i = 0
for seotag in SEOTAG:  # r: 一個li區塊
    list_seo_tag.append(seotag.getText())
    print(seotag.getText())
    print(i)
    i += 1
```

```python
# List 轉 Series
series_id = pd.Series(list_id)
series_title_h1 = pd.Series(list_title_h1)
# list_title_h2 = ",".join(list_title_h2)
series_title_h2 = pd.Series(list_title_h2)

series_content_class = pd.Series(list_content_class)
series_content_date = pd.Series(list_content_date)
series_content_all = pd.Series(list_content_all)
# list_seo_tag = ",".join(list_seo_tag)
series_seo_tag = pd.Series(list_seo_tag)
bt_column_name = ['content_id', 'content_title1', 'content_title2', 'content_class', 'content_time', 'content_all', 'content_tag']
```

```python
df_bt_1 = pd.concat([   series_id, series_title_h1, series_title_h2, series_content_class
                      , series_content_date, series_content_all, series_seo_tag], axis = 1)

df_bt_1.columns = bt_column_name
```
