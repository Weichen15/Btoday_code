# 1. Import Modules

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
import time
import pandas as pd
import datetime
```

# 2. Read Data

```python
df_bt_file = pd.read_csv('SW_Btoday_KeyWords.csv')
```

```python
df_bt_file
```

# 3. Start to Scrape and Export Data

```python
url_number = 0
today_date = datetime.date.today()
df_bt_sw_all = pd.DataFrame()

for URL in df_bt_file['URL']:
    time.sleep(5)
    url = URL
    print(url)
    headers={"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36"}
    response = requests.get(url)
    bt_html = BeautifulSoup(response.text, "lxml")
#     print(bt_html)

    # --- Begin 抓出文章id，之後會用到 ---
    
#     lst_url = df_bt_file['URL'][url_number].split('/')
        
#     for item in lst_url:
#         if len(item) == 12 and str.isdigit(item) == True:
#             content_id = item
    
#     print(content_id)
    
    content_id = bt_html.find("meta", {"property":"dable:item_id"}).get("content")
    print(content_id)
    # --- End 抓出文章id，之後會用到 ---
    
    # --- Begin 大標題 ---
    
    lst_title_h1 = []
    title_h1 = bt_html.find_all("h1")  # 搜尋所有h1
    for title in title_h1:
        print(title.getText())  # 只提取文字部分
        lst_title_h1.append(title.getText())
    
    # --- End 大標題 ---
    
    # --- Begin 次標題 ---
    
    lst_title_h2 = []
    title_h2 = bt_html.find_all("h2")  # 搜尋所有h2
    for title in title_h2:
        print(title.getText())  # 只提取文字部分
        lst_title_h2.append(title.getText())
    
    # --- End 次標題 ---
        
    # --- Begin 文章分類及文章發佈日 ---
    
    content_all = bt_html.find_all("div", {"class":"article__wrap"})  # 文章內容
    
    lst_content_class_main = []
    lst_content_class_sub = []
    lst_content_date = []
    
    # 文章屬性
    
    bt_class_main = bt_html.find("meta", {"property":"article:section"}).get("content")

    print(bt_html.find("meta", {"property":"article:section"}).get("content"))
    lst_content_class_main.append(bt_class_main)

    for content in content_all:  # r: 一個li區塊
        ct_class = content.find("p", {"class":"context__info-item context__info-item--type"})  # 文章分類
        ct_date  = content.find("p", {"class":"context__info-item context__info-item--date"})  # 文章發佈日
        lst_content_class_sub.append(ct_class.getText())
        lst_content_date.append(ct_date.getText()[:10])
        print(ct_class.getText()[:10])
        print(ct_date.getText())
        
    # --- End 文章分類及文章發佈日 ---
    
    # --- Begin 前言 ---

    lst_content_introduction = []

    for content in content_all:
        ct_introduction = content.find("div", {"class":"article__introduction"})  # 文章簡介
        lst_content_introduction.append(ct_introduction.getText())
        print(ct_introduction.getText())
    
    # --- End 前言 ---
    
    # --- Begin 內文 ---

    lst_content_all = []

    for content in content_all:  # r: 一個li區塊
        print(f"article__page__{content_id}")
        ct_all = content.find("div", {"id":f"article__page__{content_id}"})  # 文章內文 (article+page+文章ID)
#         print(ct_all)
        lst_content_all.append(ct_all.getText())
    #     print(ct_all.getText())
    
    # --- End 內文 ---
    
    # --- Begin 標籤 ---
    
    lst_content_tag = []
    content_tag = bt_html.find_all("a", {"class":"context__tag"})  # 標籤
    i = 0
    for tag in content_tag:  # r: 一個li區塊
        lst_content_tag.append(tag.getText())
        print(tag.getText())
        print(i)
        i += 1
        
    # --- End 標籤 ---
    
    
    
    # List 轉 Series
    series_id = pd.Series(content_id)
    series_title_h1 = pd.Series(lst_title_h1)
    
    lst_title_h2 = ",".join(lst_title_h2) # 把list中的資料用","join起來
    series_title_h2 = pd.Series(lst_title_h2)
    
    series_content_class_main = pd.Series(lst_content_class_main)
    series_content_class_sub = pd.Series(lst_content_class_sub)
    series_content_date = pd.Series(lst_content_date)
    series_content_all = pd.Series(lst_content_all)
    
    lst_content_tag = ",".join(lst_content_tag) # 把list中的資料用","join起來
    series_content_tag = pd.Series(lst_content_tag)
    
    bt_column_name = ['content_id'
                     ,'content_title_h1'
                     ,'content_title_h2'
                     ,'content_class_main'
                     ,'content_class_sub'
                     ,'content_date_time'
                     ,'content_all'
                     ,'content_tag']
    
    # Series 放到 DataFrame
    df_bt_sw_keyword = pd.concat([series_id
                                 ,series_title_h1
                                 ,series_title_h2
                                 ,series_content_class_main
                                 ,series_content_class_sub
                                 ,series_content_date
                                 ,series_content_all
                                 ,series_content_tag]
                                 ,axis = 1)
    
    df_bt_sw_keyword.columns = bt_column_name
    
    url_number += 1
    
    df_bt_sw_all = pd.concat([df_bt_sw_all, df_bt_sw_keyword], ignore_index=True)
    
print("----------------- Finish -------------------")

df_bt_sw_all.to_csv(f'{today_date}_BusinessToday_SimilarWeb_Keyword.csv', index = False, encoding = 'UTF-8-sig')
```
