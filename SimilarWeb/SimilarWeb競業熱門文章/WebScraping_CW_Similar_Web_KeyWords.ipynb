```python
from selenium import webdriver
import time
import os
import pandas as pd

import numpy as np
import datetime
import re
```

```python
## 匯入web資料
"""
content_id
content_title_h1
content_title_h2
content_class_main
content_class_sub
content_date_time
content_all
content_tag
"""

df_cw_file = pd.read_csv('SW_CW_KeyWords.csv')

df_cw_file
```

```python
# web 之後動態調整，讀取欲抓取網站的url

url_number = 0
today_date = datetime.date.today()
df_cw_sw_all = pd.DataFrame()
wrong_post_id = [] # 這邊不會用到

for URL in df_cw_file['URL']:
    
    web = URL

    print(web)
    try:

        # --- Begin 抓取文章id ---

        content_id = web.split("/")[-1]
        print(content_id)

        # --- End 抓取文章id ---

        path = "C:/Users/JoeJian/OneDrive - Business Today Magazine/桌面/Python Codes/Web_Scraping_in_Python_BeautifulSoup_Selenium&Scrapy_2023/ChromeDriver/chromedriver"

        driver = webdriver.Chrome(path)

        driver.get(web)

        driver.maximize_window()


        content_all_drive = driver.find_element_by_xpath(f"//article[@data-link='{content_id}']")

        # --- Begin 大標題 ---

        lst_title_h1 = []

        content_title_h1 = content_all_drive.find_element_by_xpath("./div/div/h1").text

        print(content_title_h1)

        lst_title_h1.append(content_title_h1)

        # --- End 大標題 ---

        # --- Begin 次標題 ---

        lst_title_h2 = []
        h2_title = content_all_drive.find_elements_by_xpath("./div/div/h2")

        for title in h2_title:
            print(title.text)
            lst_title_h2.append(title.text)

        # --- End 次標題 ---

        # --- Begin 文章分類及文章發佈日 ---

        lst_content_class_main = []
        lst_content_class_sub = []
        lst_content_date = []

        content_class_main = driver.find_element_by_xpath("(//ul[@class='breadcrumb list-reset']/li)[1]").text
        content_class_sub = driver.find_element_by_xpath("(//ul[@class='breadcrumb list-reset']/li)[2]").text

        print(content_class_main)
        lst_content_class_main.append(content_class_main)
        print(content_class_sub)
        lst_content_class_sub.append(content_class_sub)

        # 文章發布時間

        content_date_time = driver.find_element_by_tag_name("time").text

        content_date_time = content_date_time[-10:]

        print(content_date_time)

        lst_content_date.append(content_date_time)

        # --- End 文章分類及文章發佈日 ---

        # --- Begin 前言 ---

        lst_content_introduction = []

        content_instrunction = driver.find_element_by_xpath("//div[@class='preface']").text
        print(content_instrunction)
        lst_content_introduction.append(content_instrunction)

        # --- End 前言 ---

        # --- Begin 內文 ---

        lst_content_all = []

        content_all = content_all_drive.find_element_by_class_name("article__content").text

        lst_content_all.append(content_all)

        # --- End 內文 ---

        # --- Begin 標籤 ---

        lst_content_tag = []

        content_tag = content_all_drive.find_element_by_class_name("article__keyword").text

        lst_content_tag = content_tag.replace("#","").split("\n")

        print(lst_content_tag)

        # --- End 標籤 ---

        driver.quit()

        # List 轉 Series
        series_id = pd.Series(content_id)
        series_title_h1 = pd.Series(lst_title_h1)

        lst_title_h2 = ",".join(lst_title_h2) # 把list中的資料用","join起來
        series_title_h2 = pd.Series(lst_title_h2)

        series_content_class_main = pd.Series(lst_content_class_main)
        series_content_class_sub = pd.Series(lst_content_class_sub)
        series_content_date = pd.Series(lst_content_date)
        series_content_all = pd.Series(lst_content_all)

        lst_content_tag = ",".join(lst_content_tag) # 把list中的資料用","join起來
        series_content_tag = pd.Series(lst_content_tag)

        bt_column_name = ['content_id'
                         ,'content_title_h1'
                         ,'content_title_h2'
                         ,'content_class_main'
                         ,'content_class_sub'
                         ,'content_date_time'
                         ,'content_all'
                         ,'content_tag']

        # Series 放到 DataFrame
        df_cw_sw_keyword = pd.concat([series_id
                                     ,series_title_h1
                                     ,series_title_h2
                                     ,series_content_class_main
                                     ,series_content_class_sub
                                     ,series_content_date
                                     ,series_content_all
                                     ,series_content_tag]
                                     ,axis = 1)

        df_cw_sw_keyword.columns = bt_column_name

        url_number += 1

        df_cw_sw_all = pd.concat([df_cw_sw_all, df_cw_sw_keyword], ignore_index=True)
    
    except:
        
        driver.quit()
        
        print("Manul:", web)
        
        wrong_post_id.append(web)
        print("wrong_post_id", wrong_post_id)
        
        series_id = pd.Series(wrong_post_id)
        
        df_cw_sw_keyword_except = pd.concat([series_id], axis = 1)
        
        df_cw_sw_keyword_except.columns = ["content_id"]
        df_cw_sw_all = pd.concat([df_cw_sw_all, df_cw_sw_keyword_except], ignore_index=True)
        
        url_number += 1
    
print("----------------- Finish -------------------")

df_cw_sw_all.to_csv(f'{today_date}_CommonWealth_SimilarWeb_Keyword.csv', index = False, encoding = 'UTF-8-sig')
```
