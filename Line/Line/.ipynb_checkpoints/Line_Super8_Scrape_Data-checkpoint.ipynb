## 2. 抓文章內容的Function、抓文章點擊的Function

```python
def func_article_class(bt_link, bt_content_id, bt_title_h1, bt_title_h2, bt_class_main, bt_class_sub, bt_content_date
                       , bt_instroduction, bt_content_all, bt_tag):
    try:
        time.sleep(2)
                
        print(bt_link)
        
        headers={"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"}
        response = requests.get(bt_link, headers = headers)
        
        print("--1--", response) ## test
        
        bt_html = BeautifulSoup(response.text, "lxml")

        # --- Begin 抓出文章id，之後會用到 ---
    
        content_id = bt_html.find("meta", {"property":"dable:item_id"}).get("content")
        bt_content_id.append(str(content_id)) ## 2024/03/08 將ID型態改成string

        # --- End 抓出文章id，之後會用到 ---
        
        # --- Begin 大標題 ---
        
        h1_num = 1
        
        title_h1 = bt_html.find_all("h1")  # 搜尋所有h1
        for title in title_h1:
            print("h1:", title.getText())  # 只提取文字部分
            
            if h1_num == 1:
                bt_title_h1.append(title.getText())
            
            h1_num += 1
        # --- End 大標題 ---
        
        ## ---- 區分幸福熟齡以及今周刊 ----
        
        if "幸福熟齡" in bt_title_h1[0]:
            
            print("OH YEAH")
            
            content_all = bt_html.find("div", {"itemprop":"articleBody"})

            # --- Begin 次標題 ---
            bt_title_h2_temp = []

            title_h2 = content_all.find_all("h2")  # 搜尋所有h2
            for title in title_h2:
        #             print("h2:", title.getText())  # 只提取文字部分
                bt_title_h2_temp.append(title.getText())
                print(title.getText())
            print(bt_title_h2_temp)

            # --- End 次標題 ---
            
            # --- Begin 文章分類 ---

            ct_class_main = bt_html.find("meta", {"property":"article:section"}).get("content")

            print(bt_html.find("meta", {"property":"article:section"}).get("content"))
            bt_class_main.append(ct_class_main)

            ct_class_sub = bt_html.find("meta", {"property":"article:section2"}).get("content")

            print(bt_html.find("meta", {"property":"article:section2"}).get("content"))
            bt_class_sub.append(ct_class_sub)

            # --- End 文章分類 ---

            # --- Begin 文章發佈日 ---

            ct_date = bt_html.find("meta", {"property":"article:published_time"}).get("content")
            bt_content_date.append(ct_date[:10])
            print(bt_content_date)

            # --- End 文章分類及文章發佈日 ---

            # --- Begin 前言 ---

            ct_introduction = bt_html.find("meta", {"itemprop":"description"}).get("content")
            # 文章簡介
            bt_instroduction.append(ct_introduction)
            print(bt_instroduction)

            # --- End 前言 ---

            # --- Begin 內文 ---


            ct_all = bt_html.find("div", {"itemprop":"articleBody"})  # 文章內文 (article+page+文章ID)
            bt_content_all.append(ct_all.getText())
            print(bt_content_all)

            # --- End 內文 ---

            # --- Begin 標籤 ---


            bt_tag_temp = []
            content_tag = bt_html.find_all("ul", {"class":"article_tags flex items-center mb-5 flex-wrap"})  # 標籤  # 標籤

            for tag in content_tag:  # r: 一個li區塊
                bt_tag_temp.append(tag.getText())
                print(tag.getText())

            # --- End 標籤 ---
            
        else: ## ---- 今周刊 ---- ##

            # --- Begin 次標題 ---
            bt_title_h2_temp = []

            title_h2 = bt_html.find_all("h2")  # 搜尋所有h2
            for title in title_h2:
    #             print("h2:", title.getText())  # 只提取文字部分
                bt_title_h2_temp.append(title.getText())

            # --- End 次標題 ---

            # --- Begin 文章分類 ---

            ct_class_main = bt_html.find("meta", {"property":"article:section"}).get("content")

            print(bt_html.find("meta", {"property":"article:section"}).get("content"))
            bt_class_main.append(ct_class_main)

            ct_class_sub = bt_html.find("meta", {"property":"article:section2"}).get("content")

            print(bt_html.find("meta", {"property":"article:section2"}).get("content"))
            bt_class_sub.append(ct_class_sub)

            # --- End 文章分類 ---

            # --- Begin 文章發佈日 ---

            content_all = bt_html.find_all("div", {"class":"article__wrap"})  # 文章內容


            for content in content_all:  # r: 一個li區塊
                ct_date  = content.find("p", {"class":"context__info-item context__info-item--date"})  # 文章發佈日
                bt_content_date.append(ct_date.getText()[:10])

            # --- End 文章分類及文章發佈日 ---

            # --- Begin 前言 ---


            for content in content_all:
                ct_introduction = content.find("div", {"class":"article__introduction"})  # 文章簡介
                bt_instroduction.append(ct_introduction.getText())

            # --- End 前言 ---

            # --- Begin 內文 ---


            for content in content_all:  # r: 一個li區塊

                ct_all = content.find("div", {"id":f"article__page__{content_id}"})  # 文章內文 (article+page+文章ID)
                bt_content_all.append(ct_all.getText())

            # --- End 內文 ---
            
        

            # --- Begin 標籤 ---


            bt_tag_temp = []
            content_tag = bt_html.find_all("a", {"class":"context__tag"})  # 標籤

            for tag in content_tag:  # r: 一個li區塊
                bt_tag_temp.append(tag.getText())

            # --- End 標籤 ---
        
        ## ---- 區分幸福熟齡以及今周刊 ----

        bt_title_h2_temp = ",".join(bt_title_h2_temp) # 把list中的資料用","join起來
        bt_title_h2.append(bt_title_h2_temp)
#         print("bt_title_h2:", bt_title_h2)

        bt_tag_temp = ",".join(bt_tag_temp) # 把list中的資料用","join起來
        bt_tag.append(bt_tag_temp)
#         print("bt_tag:", bt_tag)

        
    except:
        pass
        print("nothing")
        
        bt_content_id.append(bt_link)
        bt_title_h1.append(bt_link)
        bt_title_h2.append(bt_link)
        bt_class_main.append(bt_link)
        bt_class_sub.append(bt_link)
        bt_content_date.append(bt_link)
        bt_instroduction.append(bt_link)
        bt_content_all.append(bt_link)
        bt_tag.append(bt_link)
```

```python
def func_merge_click_data(list_click, list_click_rate, list_click_user, list_click_user_rate
                          , btn_num, df_data, post_data_index):
    
    click_btn_num_1 = 15 + btn_num # 點擊次數
#     click_btn_num_1 = 26  # 點擊次數
    click_btn_num_2 = 17 + btn_num #重複點擊率
    click_btn_num_3 = 16 + btn_num # 點擊人數
    click_btn_num_4 = 18 + btn_num #不重複點擊率
    
    list_click.append(df_data.iloc[(post_data_index, click_btn_num_1)])
    list_click_rate.append(df_data.iloc[(post_data_index, click_btn_num_2)])
    list_click_user.append(df_data.iloc[(post_data_index, click_btn_num_3)])
    list_click_user_rate.append(df_data.iloc[(post_data_index, click_btn_num_4)])
    
    print("點擊次數: ", df_data.iloc[(post_data_index, click_btn_num_1)]) # 1換成df_post_data的第幾列
    print("重複點擊率: ", df_data.iloc[(post_data_index, click_btn_num_2)])
    print("點擊的用戶: ", df_data.iloc[(post_data_index, click_btn_num_3)])
    print("不重複點擊率: ", df_data.iloc[(post_data_index, click_btn_num_4)])
```

### 資料爬取function

```python
def super8_data(df_post_data, post_ad):
    """
    df_post_data: 放內容組發文資料或者其他
    post_ad:輸出資料時，檔名為post或ad
    
    """
    ### --- 調整發送時間，切成日期、星期幾、時間 start ---

    lst_post_date = []
    lst_post_weekday = []
    lst_post_time = []
    lst_post_date_time = []

    for date_time in df_post_data["發送時間"]:

        lst_post_date_time.append(date_time)

        lst_post_date.append(date_time.split(" ")[0]) ## 發文日期   
        lst_post_time.append(date_time.split(" ")[1]) ## 發文時間

        post_year = int(date_time.split(" ")[0].split('/')[0])
        post_month = int(date_time.split(" ")[0].split('/')[1])
        post_date = int(date_time.split(" ")[0].split('/')[2])

        today_date = datetime.datetime(post_year, post_month, post_date)
        today_day = today_date.strftime("%w")
        if today_day == "0":
            today_day = "7"

        lst_post_weekday.append(today_day) ## 發文星期
    ### --- 調整發送時間，切成日期、星期幾、時間 end ---


    ### --- 篩選所需欄位 start ---
    df_post_date_time = pd.DataFrame({"發文日期": lst_post_date
                                     ,"發文星期": lst_post_weekday
                                     ,"發文時間": lst_post_time
                                     ,"Post_date_time": lst_post_date_time})
    
    ### --- 篩選所需欄位 end ---

    ### --- 合併dataframe start ---
    df_post_data_1 = df_post_data.reset_index(drop = True) # reset index for concat

    df_post_data_datetime = pd.concat([df_post_date_time, df_post_data_1], axis = 1)
    
    ### --- 合併dataframe end ---

    ### --- 開始爬取資料 start ---
    df_line_super8_all = pd.DataFrame() ## 每則貼文資訊
    df_post_broadcast_data = pd.DataFrame() ## 單一推播資訊

    for post_num in range(len(df_post_data)):
        post_name = df_post_data.iloc[(post_num, 1)] # 之後用讀資料方式抓取
        print("post_name", post_name)


        # post name
    #     time.sleep(2.5)
        post_name_search = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//input[@type='text']")))

    #     post_name_search = driver.find_element_by_xpath("//input[@type='text']")
        post_name_search.send_keys(post_name)
        post_name_search.send_keys(Keys.RETURN)

        ## ----- 點進去該則訊息 ----- ##

    #     time.sleep(2)

        ## ----- 特殊例外 ----- ##
    #     print("Ori post_name: ", post_name)
    #     if post_name == '申霖_周日LINE圖_2312032100':
    #         print('skip')
    #         post_list_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[@id='依榕_周六LINE圖_2311251500'][2]")))
    #         post_name = '慧如_週五金鐘LINE圖_2310201800(重發)'
    #         print("New post_name: ", post_name)
    #     else:
    #         post_list_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, post_name)))
    #         post_list_button.click()
        ## ----- 特殊例外 ----- ##

        time.sleep(2)
        post_list_button = WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.ID, post_name)))

        try:
            post_list_button.click()

        except StaleElementReferenceException: # After encountering a stale element reference, you can refresh the entire page and try locating the element again.
            driver.refresh()
            post_list_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, post_name)))
            post_list_button.click()

    #     post_list_button = driver.find_element_by_id(post_name)
    #     time.sleep(5)
    #     post_list_button.click()

        ## ----- 推播通知訊息 ----- ##

    #     time.sleep(3)

        lst_broadcast_info_message = []

        broadcast_info_message = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//*[@id='root']/div/main/div/div[2]/div/div[6]/div[11]"))).text

    #     broadcast_info_message = driver.find_element_by_xpath("//*[@id='root']/div/main/div/div[2]/div/div[6]/div[11]").text
        print("推播通知訊息:", broadcast_info_message)
        lst_broadcast_info_message.append(broadcast_info_message)
        print(lst_broadcast_info_message)

        ## ----- 推播訊息 ----- ##
        broadcast_all = driver.find_element_by_xpath("//*[@id='root']/div/main/div/div[2]/div/div[10]").text

        broadcast_all_list = broadcast_all.split("\n")

        lst_broadcast_type = []
        lst_broadcast_link = []

        for broadcast_num in range(len(broadcast_all_list)):

            message_index = broadcast_num - 7

            if (message_index) % 11 == 0:
                print("test1: ", message_index, broadcast_all_list[broadcast_num])
                lst_broadcast_type.append(f'圖{broadcast_all_list[broadcast_num]}')

            if (message_index) % 11 == 3:
                print("test2: ", message_index, broadcast_all_list[broadcast_num])
                lst_broadcast_link.append(broadcast_all_list[broadcast_num])

        print("type: ", lst_broadcast_type)
        print("link: ", lst_broadcast_link)

        ## ----- 抓取官網資料 ----- ##
        ### --- create list for article class ---
        lst_content_id = []
        lst_title_h1 = []
        lst_title_h2 = []
        lst_article_class_main = []
        lst_article_class_sub = []
        lst_content_date = []
        lst_content_introduction = []
        lst_content_all = []
        lst_content_tag = []
        # --- --- --- --- --- --- --- --- --- ---

        for article_link in lst_broadcast_link:
            func_article_class( article_link, lst_content_id, lst_title_h1, lst_title_h2, lst_article_class_main, lst_article_class_sub
                                  , lst_content_date, lst_content_introduction, lst_content_all, lst_content_tag)
        # 抓取今周刊官網資料

        ## ----- 合併下載檔案資料 ----- ##


        lst_click = [] # 點擊次數 or 點擊數
        lst_click_rate = [] # 重複點擊率 or 點擊率
        lst_click_user = [] # 點擊人數 or 點擊的用戶
        lst_click_user_rate = [] # 不重複點擊率

        for post_type in lst_broadcast_type:
            if post_type == "圖1":
                btn_num = 0
                func_merge_click_data(lst_click, lst_click_rate, lst_click_user, lst_click_user_rate
                                     ,btn_num, df_post_data, post_num)

            elif post_type == "圖2":
                btn_num = 4
                func_merge_click_data(lst_click, lst_click_rate, lst_click_user, lst_click_user_rate
                                     ,btn_num, df_post_data, post_num)

            else:
                btn_num = 8
                func_merge_click_data(lst_click, lst_click_rate, lst_click_user, lst_click_user_rate
                                     ,btn_num, df_post_data, post_num)

        df_link_click_temp = pd.DataFrame({'連結網址': lst_broadcast_link
                                          ,'發文類型': lst_broadcast_type
                                          ,'點擊數': lst_click
                                          ,'重複點擊率': lst_click_rate
                                          ,'點擊人數': lst_click_user
                                          ,'不重複點擊率': lst_click_user_rate
                                          ,'文章ID': lst_content_id
                                          ,'文章標題_h1': lst_title_h1
                                          ,'文章標題_h2': lst_title_h2
                                          ,'文章大分類': lst_article_class_main
                                          ,'文章次分類': lst_article_class_sub
                                          ,'文章發布日': lst_content_date
                                          ,'文章前言': lst_content_introduction
                                          ,'文章全文': lst_content_all
                                          ,'文章標籤': lst_content_tag
                                           })

    #     print("df_link_click_temp", df_link_click_temp)

        df_post_status_temp = df_post_data_datetime[["發文日期", "發文星期", "發文時間", "Post_date_time", '群發訊息名稱', '群發最後建立者'
                         , '發送成功人數', '開封率', '點擊率','未開封人數', '已開封人數', '已開封未點擊人數', '已點擊人數']]

        df_post_status_temp = df_post_status_temp.iloc[[post_num]].reset_index(drop = True) ## 抓取報表有的資料

        df_broadcast_info_message = pd.DataFrame({"推播通知訊息": lst_broadcast_info_message}) ## 推播通知

        df_post_status_temp = pd.concat([df_post_status_temp, df_broadcast_info_message], axis = 1)

        df_post_broadcast_data = pd.concat([df_post_broadcast_data, df_post_status_temp] , ignore_index = True) ## 推播資料

        df_post_status_dup = pd.concat([df_post_status_temp] * len(df_link_click_temp), ignore_index = True)

        df_line_super8 = pd.concat([df_post_status_dup, df_link_click_temp], axis = 1)

        df_line_super8_all = pd.concat([df_line_super8_all, df_line_super8], ignore_index = True) ## 所有資料

        ## ----- 回到訊息列表頁 ----- ##

        back_to_list_button = driver.find_element_by_xpath("(//a[@class='MuiTypography-root MuiLink-root MuiLink-underlineHover MuiTypography-colorInherit'])[1]")
        back_to_list_button.click()
    
    if post_ad == "ad":
        df_line_super8_all = df_line_super8_all[df_line_super8_all.columns[:20]]
    
    df_line_super8_all.to_csv(f"Line_super8_{post_ad}_{file_name}.csv", index = False, encoding = 'UTF-8-sig')
    df_post_broadcast_data.to_csv(f"Line_super8_{post_ad}_broadcast_{file_name}.csv", index = False, encoding = 'UTF-8-sig')


    print(f"------------- {post_ad} Finish --------------")
    
### --- 開始爬取資料 end ---
```
