## 2. æŠ“Poæ–‡é»æ“Šçš„Function

```python
def func_merge_click_data(list_click, list_click_rate, list_click_user, list_click_user_rate
                          , btn_num, df_data, post_data_index):
    
    click_btn_num_1 = 15 + btn_num # é»æ“Šæ¬¡æ•¸
#     click_btn_num_1 = 26  # é»æ“Šæ¬¡æ•¸
    click_btn_num_2 = 17 + btn_num #é‡è¤‡é»æ“Šç‡
    click_btn_num_3 = 16 + btn_num # é»æ“Šäººæ•¸
    click_btn_num_4 = 18 + btn_num #ä¸é‡è¤‡é»æ“Šç‡
    
    list_click.append(df_data.iloc[(post_data_index, click_btn_num_1)])
    list_click_rate.append(df_data.iloc[(post_data_index, click_btn_num_2)])
    list_click_user.append(df_data.iloc[(post_data_index, click_btn_num_3)])
    list_click_user_rate.append(df_data.iloc[(post_data_index, click_btn_num_4)])
    
    print("é»æ“Šæ¬¡æ•¸: ", df_data.iloc[(post_data_index, click_btn_num_1)]) # 1æ›æˆdf_post_dataçš„ç¬¬å¹¾åˆ—
    print("é‡è¤‡é»æ“Šç‡: ", df_data.iloc[(post_data_index, click_btn_num_2)])
    print("é»æ“Šçš„ç”¨æˆ¶: ", df_data.iloc[(post_data_index, click_btn_num_3)])
    print("ä¸é‡è¤‡é»æ“Šç‡: ", df_data.iloc[(post_data_index, click_btn_num_4)])
```

```python
df_super8_data = pd.read_csv(f"Super8_{file_name}.csv", encoding = 'big5')

df_super8_data.head(5)
```

```python
line_ad_post = (df_super8_data["ç™¼é€å…§å®¹"] == "å¤šå‰‡è¨Šæ¯") & (df_super8_data["ç™¼é€æˆåŠŸäººæ•¸"] > 1)

df_ad_data = df_super8_data.loc[line_ad_post]

df_ad_data.head(5)
```

### èª¿æ•´ç™¼é€æ™‚é–“ï¼Œåˆ‡æˆæ—¥æœŸã€æ˜ŸæœŸå¹¾ã€æ™‚é–“

```python
lst_post_date = []
lst_post_weekday = []
lst_post_time = []
lst_post_date_time = []

for date_time in df_ad_data["ç™¼é€æ™‚é–“"]:
    
    lst_post_date_time.append(date_time)
    
    lst_post_date.append(date_time.split(" ")[0]) ## ç™¼æ–‡æ—¥æœŸ   
    lst_post_time.append(date_time.split(" ")[1]) ## ç™¼æ–‡æ™‚é–“
    
    post_year = int(date_time.split(" ")[0].split('/')[0])
    post_month = int(date_time.split(" ")[0].split('/')[1])
    post_date = int(date_time.split(" ")[0].split('/')[2])
    
    today_date = datetime.datetime(post_year, post_month, post_date)
    today_day = today_date.strftime("%w")
    if today_day == "0":
        today_day = "7"

    lst_post_weekday.append(today_day) ## ç™¼æ–‡æ˜ŸæœŸ 
    
print(lst_post_date)
print(lst_post_weekday)
print(lst_post_time)
print(lst_post_date_time)
```

```python
df_post_date_time = pd.DataFrame({"ç™¼æ–‡æ—¥æœŸ": lst_post_date
                                 ,"ç™¼æ–‡æ˜ŸæœŸ": lst_post_weekday
                                 ,"ç™¼æ–‡æ™‚é–“": lst_post_time
                                 ,"Post_date_time": lst_post_date_time})

df_post_date_time.head(5)
```

```python
df_post_data_1 = df_ad_data.reset_index(drop = True) # reset index for concat

df_post_data_datetime = pd.concat([df_post_date_time, df_post_data_1], axis = 1)

df_post_data_datetime.head(5)
```
