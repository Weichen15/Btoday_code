```python
# Define get class function


def func_article_class(bt_link, bt_class):
    try:
        # -- Begin Get BT URL --

        driver_bt = webdriver.Chrome(g_path)

        driver_bt.get(bt_link)

        driver_bt.maximize_window()

        # -- End Get BT URL --

        # -- Begin Find Article Class

        link_class = driver_bt.find_element_by_xpath("//p[@class='context__info-item context__info-item--type']").text
        print(link_class)

        bt_class.append(link_class)

#         # -- Quit BT Virtual Web --
#         driver_bt.quit()
    except:
        pass
    
    # -- Quit BT Virtual Web --
    driver_bt.quit()
```

```python
# Define type Link scraping function

def func_link_class(fb_link, lst_class):
    # first arg: Facebook link that you want to scrape
    # second arg: List that you append classfication
    
    
    
    try:
       
        # -- Begin Get FB URL --

        web_fb = fb_link

        driver_fb = webdriver.Chrome(g_path)

        driver_fb.get(web_fb)

        driver_fb.maximize_window()

        # -- End Get FB URL --

        # -- Begin Find to BT Web Link in FB --

        web_bt_all = WebDriverWait(driver_fb, 5).until(EC.presence_of_element_located((By.XPATH, "//div[@class='_3x-2']")))
        web_bt = web_bt_all.find_element_by_xpath("//div[@class='_6ks']/a").get_attribute('href')
#         web_bt = driver_fb.find_element_by_xpath("//div[@class='_6ks']/a").get_attribute('href')
  
#         print('web_bt', web_bt)
    
        # -- End Find to BT Web Link in FB --

        # -- Quit FB Virtual Web --
#         time.sleep(2)
        driver_fb.quit()
        
        if 'thebetteraging' in (web_bt):
#             print(web_bt)

            link_class = '幸福熟齡'
            print(link_class)

            lst_class.append(link_class)

        elif 'esg.businesstoday' in (web_bt):

            link_class = 'ESG'
            print(link_class)

            lst_class.append(link_class)

        else:
            
            func_article_class(web_bt, lst_class)
#             # -- Begin Get BT URL --

#             driver_bt = webdriver.Chrome(g_path)

#             driver_bt.get(web_bt)

#             driver_bt.maximize_window()

#             # -- End Get BT URL --

#             # -- Begin Find Article Class

#             link_class = driver_bt.find_element_by_xpath("//p[@class='context__info-item context__info-item--type']").text
#             print(link_class)

#             lst_class.append(link_class)

#             # -- Quit BT Virtual Web --
#             driver_bt.quit()
        
    except:
        lst_class.append(fb_link)
        
    # -- Quit FB Virtual Web --

    driver_fb.quit()
```

```python
# Define type Text scraping function

def func_text_class(fb_link, lst_class):
    # first arg: Facebook link that you want to scrape
    # second arg: List that you append classfication
    
    try:
       
        # -- Begin Get FB URL --

        web_fb = fb_link

        driver_fb = webdriver.Chrome(g_path)

        driver_fb.get(web_fb)

        driver_fb.maximize_window()

        # -- End Get FB URL --

        # -- Begin Click comment button --
#         time.sleep(2)
        comment_button = WebDriverWait(driver_fb, 5).until(EC.presence_of_element_located((By.XPATH, "//span[@class='_1whp _4vn2']")))
#         comment_button = driver_fb.find_element_by_xpath("//span[@class='_1whp _4vn2']")

        comment_button.click()
        
        # -- End Click comment button -- 
        
        # -- Begin Get article link --
        web_bt_all = WebDriverWait(driver_fb, 5).until(EC.presence_of_element_located((By.XPATH, "//div[@class='_4eeo']")))
        web_bt = web_bt_all.find_element_by_xpath("//div[@class='_3mvv']/a").get_attribute('href')
        
#         web_bt = driver_fb.find_element_by_xpath("//div[@class='_3mvv']/a").get_attribute('href')
        
#         print('web_bt', web_bt)
                
        # -- End Get article link --

        # -- Quit FB Virtual Web --
#         time.sleep(2)
#         driver_fb.quit()
            
        if 'thebetteraging' in (web_bt):

            link_class = '幸福熟齡'
            print(link_class)

            lst_class.append(link_class)

        elif 'esg.businesstoday' in (web_bt):

            link_class = 'ESG'
            print(link_class)

            lst_class.append(link_class)

        else:
            
            func_article_class(web_bt, lst_class)
            
#             # -- Begin Get BT URL --

#             driver_bt = webdriver.Chrome(g_path)

#             driver_bt.get(web_bt)

#             driver_bt.maximize_window()

#             # -- End Get BT URL --

#             # -- Begin Find Article Class

#             link_class = driver_bt.find_element_by_xpath("//p[@class='context__info-item context__info-item--type']").text
#             print(link_class)

#             lst_class.append(link_class)

#             # -- Quit BT Virtual Web --
#             driver_bt.quit()

        driver_fb.quit()
        
    except:
        
        print("except")
        
        lst_class.append(fb_link)
        
        driver_fb.quit()
```

```python
# create list for article class
lst_article_class_Ori = []
```

```python
for link, post in zip(lst_link, lst_type):
    if post == 'Link':
#         print(f'{post}, Link')
        func_link_class(link, lst_article_class_Ori)
#         print(lst_article_class_Ori)
    elif post == 'Text':
#         print(f'{post}, Link')
        func_text_class(link, lst_article_class_Ori)
#         print(lst_article_class_Ori)
    elif post == 'Photo':
#         print(f'{post}, Photo')
        lst_article_class_Ori.append('圖片')
    elif post == 'Video':
#         print(f'{post}, Photo')
        lst_article_class_Ori.append('影音')
    else:
        print('other')
    
# print(lst_article_class_Ori)
print("Finish")
```

```python
lst_article_class_for_fb = []
```

```python
# Define Business Today Classification
lst_產業動態 = ['產業動態', '科技', '傳產', '產業動態', '國際瞭望', '兩岸三地']
lst_金融保險 = ['金融保險', '金融', '保險', '稅制', '勞保勞退']
lst_國際總經 = ['國際總經']
lst_政治社會 = ['政治社會']
lst_投資理財 = ['投資理財', '台股', '全球股市', '債券基金', '信用卡', '理財', 'ETF', '聰明理財', '股債前線', '基金投資', '房地產掃瞄', '保險稅務']
lst_健康休閒 = ['健康休閒', '健康', '健康']
lst_生活消費 = ['生活消費', '美食旅遊', '品味收藏', '情感關係', '品味人生', '美食旅遊']
lst_職場學習 = ['職場學習', '職場', '教育', '退休', '個人成長', '管理', '行銷', '創業', '教育']
lst_家庭教育 = ['家庭教育']
lst_幸福熟齡 = ['幸福熟齡', '預約美好告別']
lst_房地產   = ['房地產']
lst_ESG      = ['ESG']
lst_訂閱制   = ['訂閱制']
lst_財經時事 = ['財經時事', '焦點新聞', '科技線上', '金融風雲']
lst_今選頻道 = ['今選頻道', '話題人物', '最新觀點', '名人專欄', '編輯室報告']
```

```python
for item in lst_article_class_Ori:
    if item in lst_產業動態:
        lst_article_class_for_fb.append('產業動態')
        
    elif item in lst_金融保險:
        lst_article_class_for_fb.append('金融保險')
        
    elif item in lst_國際總經:
        lst_article_class_for_fb.append('國際總經')
        
    elif item in lst_政治社會:
        lst_article_class_for_fb.append('政治社會')
        
    elif item in lst_投資理財:
        lst_article_class_for_fb.append('投資理財')
        
    elif item in lst_健康休閒:
        lst_article_class_for_fb.append('健康休閒')

    elif item in lst_生活消費:
        lst_article_class_for_fb.append('生活消費')
        
    elif item in lst_職場學習:
        lst_article_class_for_fb.append('職場學習')
        
    elif item in lst_家庭教育:
        lst_article_class_for_fb.append('家庭教育')
        
    elif item in lst_幸福熟齡:
        lst_article_class_for_fb.append('幸福熟齡')
        
    elif item in lst_房地產:
        lst_article_class_for_fb.append('房地產')
        
    elif item in lst_ESG:
        lst_article_class_for_fb.append('ESG')
        
    elif item in lst_訂閱制:
        lst_article_class_for_fb.append('訂閱制')
        
    elif item in lst_今選頻道:
        lst_article_class_for_fb.append('今選頻道')
        
    else:
        lst_article_class_for_fb.append(item)
```
